# 1- Shapiro Wilk Test
Test whether a data has a Gaussian Distribution
 - **Assumption**
   - Observation in each sample are independent and identically distributed(iid)
   - Interpretation:
     - **Ho:** The sample has a Gaussian Distribution
     - **H1:** The sample does not have a Gaussian Distribution

# 3- Parameteric Statistical Hypothesis Tests

### 1- Student t-test

Test whether the mean of two independent samples are significantly different 

**Assumption:** 
 - Observation in each sample are independent and identically different 
 - Observation in each sample are normally distributed 
 - Observation in each sample have the same variance
 - Interpretation:
   - **Ho:** The means of the samples are equal
   - **H1:** The means of the samples are unequal

```python
# install the essential modules
pip install numpy
pip install pandas 
pip install matplotlib
pip install seaborn 
pip install scipy
```

```python
# import essential libraries 
import numpy as np 
import seaborn as sns  
import matplotlib.pyplot as plt 
import pandas as pd  
from scipy.stats import ttest_ind

# Load the data set 
data1=[0.873, 2.817, 0.121, -0.945, -0.055, -1.436, 0.360, -1.478, -1.637, -1.869]
data2=[0.353, 3.517, 0.125, -7.545, -0.555, -1.536, 3.350, -1.578, -3.537, -1.579]
```

```python
# chi squared test
stat, p=ttest_ind(data1,data2)
print("stat=%.3f, p=%.3f" % (stat,p))
```

```python
stat, p=ttest_ind(data1,data2)
if p>0.05:
    print("Probably Independent (No correlation)")
else:
    print("Probably dependent (Correlation exist)")
```

# Analysis of Variance (ANOVA)

Test whether the mean of two or more independent samples are significantly different 

**Assumption:** 
 - Observation in each sample are independent and identically different (iid) 
 - Observation in each sample are normally distributed 
 - Observation in each sample have the same variance
 - Interpretation:
   - **Ho:** The means of the samples are equal
   - **H1:** The means of the samples are unequal

```python
# install the essential modules
pip install numpy
pip install pandas 
pip install matplotlib
pip install seaborn 
pip install scipy
```

```python
# import essential libraries 
import numpy as np 
import seaborn as sns  
import matplotlib.pyplot as plt 
import pandas as pd 
from scipy.stats import f_oneway 


# Load the data set 
data1=[0.873, 2.817, 0.121, -0.945, -0.055, -1.436, 0.360, -1.478, -1.637, -1.869]
data2=[0.353, 3.517, 0.125, -7.545, -0.555, -1.536, 3.350, -1.578, -3.537, -1.579]
data3=[0.353, 3.517, 0.125, -7.545, -0.055, -1.436, 0.360, -1.478, -1.637, -1.869]
```

```python
stat,p=f_oneway(data1,data2,data3)
print("stat=%.3f, p=%.3f" % (stat,p))
```

```python
stat,p=f_oneway(data1,data2,data3)
if p>0.05:
    print("Probably Independent (No correlation)")
else:
    print("Probably dependent (Correlation exist)")
```
